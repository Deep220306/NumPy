Advanced Array Manipulation 

Stacking Array

```python
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
```

```python
np.vstack((a, b)) # Vertical stacking
```

```python
np.hstack((a, b)) # Horizontal stacking
```

```python
np.column_stack((a, b)) # Column stacking
```

Splitting array 

```python
c = np.arange(16).reshape((4, 4))
c
```

```python
np.hsplit(c, 2) # Split c into 2 equal arrays horizontally
```

```python
np.vsplit(c, 2) # Split c into 2 equal arrays vertically
```
